```python
def Crypto_cleaning_newdata(data_crypto_two):
    data_crypto_two["max_supply"] = data_crypto_two["max_supply"].fillna(0).astype(int)
    data_crypto_two["circulating_supply"]= data_crypto_two["circulating_supply"].fillna(0).astype(int)
    data_crypto_two["total_supply"]= data_crypto_two["total_supply"].fillna(0).astype(int)
    data_crypto_two["quote.USD.market_cap"]= data_crypto_two["quote.USD.market_cap"].fillna(0).astype(int)
    data_crypto_two["quote.USD.fully_diluted_market_cap"]= data_crypto_two["quote.USD.fully_diluted_market_cap"].fillna(0).astype(int)
    data_crypto_two["date_added"]= pd.to_datetime( data_crypto_two["date_added"],infer_datetime_format=True)
    data_crypto_two["quote.USD.last_updated"] = pd.to_datetime( data_crypto_two["quote.USD.last_updated"],infer_datetime_format=True)
    return data_crypto_two
```
